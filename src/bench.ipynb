```python
import matplotlib.pyplot as plt
```

```python
# draw a plot to show data 
data_x = [r'$2^1$', r'$2^2$', r'$2^3$', r'$2^4$', r'$2^5$', r'$2^6$', r'$2^7$']
data_y_1 = [1.5, 2.30, 3.1, 5.5, 10.4, 19.8, 39.4]
data_y_2 = [0.52, 0.55, 0.5, 0.51, 0.52, 0.52, 0.55]

plt.plot(data_x, data_y_1, 'ro-', label='proving')
plt.plot(data_x, data_y_2, 'go-', label='verifying')
plt.legend()
plt.xlabel('Number of transactions per proof')
plt.ylabel('Time (in seconds)')
plt.title("Proving and verifying time")
plt.show()
```

```python
data_x = [r'$2^1$', r'$2^2$', r'$2^3$', r'$2^4$', r'$2^5$', r'$2^6$', r'$2^7$']
data_y_1 = [1.417510271, 5.98235178, 13.23469186, 27.39548469, 60.408, 112.797465, 214.667317]
data_y_2 = [20.44802094, 24.93711567, 32.52978206, 46.5848434, 78.9655, 131.1666, 233.1598322]
data_y_3 = [0.52, 0.55, 0.5, 0.51, 0.52, 0.52, 0.55]
plt.plot(data_x, data_y_1, 'ro-', label='merging')
plt.plot(data_x, data_y_2, 'go-', label='proving')
plt.plot(data_x, data_y_3, 'yo-', label='verifying')
plt.legend(),
plt.xlabel('Number of Proofs')
plt.ylabel('Time (in seconds)')
plt.title("Proving time")
plt.show()
```
